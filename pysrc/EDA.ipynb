```python
short_match = df_match[df_match['foulcommit'].notna()][['date','home_team_api_id', 'away_team_api_id', 'foulcommit']]
short_team = df_team[['team_api_id', 'team_long_name']]
m1 = pd.merge(short_match, short_team, how = 'inner', left_on = 'home_team_api_id', right_on = 'team_api_id', suffixes = ("_x", "_y"))

m2 = pd.merge(m1, short_team, how = 'inner', left_on = 'away_team_api_id', right_on = 'team_api_id', suffixes = ('_x', '_y'))
```

```python
fouls_teams = m2.rename(
    columns = {'team_long_name_x' : 'home_team_name', 'team_long_name_y' : 'away_team_name'}
).filter(regex = 'date|home|away|foul')
```

```python
fouls_teams.head()
```

```python
import lxml.etree as etree
```

```python
xml = fouls_teams['foulcommit'].iloc[0]
root = etree.fromstring(xml)
```

```python
print(etree.tostring(root, pretty_print = True).decode())
```

```python
unique_elems = []
for child in root:
    if child[0][0].tag not in unique_elems:
        unique_elems += [child[0][0].tag]
    
    for child1 in child[1:]:
        if child1.tag not in unique_elems:
            unique_elems += [child1.tag]
unique_elems

d = {'foul_idx' : []}
for elem in unique_elems:
    d[elem] = []
    
for i, child in enumerate(root):
    d['foul_idx'] += [i]
    d[child[0][0].tag] += [child[0][0].text]
    
    for child1 in child[1:]:
        d[child1.tag] += [child1.text]
    
    key_lengths = [len(d[x]) for x in list(d.keys())]
    for x in list(d.keys()):
        if len(d[x]) < max(key_lengths):
            d[x] += [None]

pd.DataFrame(d)
```
